```python
# prompt: show gradio deployment for the above

import gradio as gr

def classify_text(text):
  """
  Classifies the input text using a pre-trained text classification model.
  """
  result = classifier(text)
  return result

iface = gr.Interface(
    fn=classify_text,
    inputs=gr.Textbox(placeholder="Enter text here..."),
    outputs=gr.Textbox(),
    title="Text Classification",
    description="Classify text using a pre-trained model.",
)

iface.launch(debug=True)
```
